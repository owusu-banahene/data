```python
x = np.arange(-10,11)
```

```python
x
```

```python
plt.figure(figsize=(12,6))
plt.title('My nice plot')
plt.plot(x,x**2)
plt.plot(x,-1*(x**2))
```

```python
fig,axes=plt.subplots(figsize=(12,6))
```

```python
axes.plot(x,x**2)
axes.plot(x,-1*(x**2))
axes.legend(["X","-X"])
axes.set_xlabel('X values')
axes.set_ylabel('X squared')
fig
```
